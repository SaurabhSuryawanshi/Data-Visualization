```python
ds2=sb.load_dataset('car_crashes')
#loading a car crashes dataset
```

```python
ds2
```

```python
ds2.plot.bar()
#plotting a bar chart
```

```python
ds2['alcohol'].plot.hist()
```

```python
ds2['speeding'].plot.hist()
#showing the dataset in histogram of particular data column
```

```python
ds2.plot(kind='hist',stacked=True,bins=20,figsize=(5,10),title='DV of car crashes')
#data stacked above one another
```

```python
ds2.plot()
#plotting the overall data in one chart
```

```python
ds2.hist(figsize=(10,15),color='#119900',alpha=.5,sharex=False)
#here you can see the frequency and count of all the data column wise
```

```python
ds2.plot.scatter(x='alcohol',y='speeding')
#scattering the data of variables who were speeding while intoxicated
```

```python
ds2.plot.scatter(x='alcohol',y='speeding',c='not_distracted')
#more the alcohol intake higher the spedding so more the distraction
```

```python
ds2.plot.scatter(x='alcohol',y='speeding',c='not_distracted',label='do not drink&drive',s=30)
#just labelling it and drawing a size to the variables
```

```python
ds2.plot.hexbin(x='alcohol',y='speeding',gridsize=30,label='HexaShape')
```

```python
ds2.plot(kind='pie', y='speeding')
#plotting a pie chart for spedding value
```

```python
ds2.plot(kind='pie',y='alcohol')
#plotting a pie chart for alcohol
```

```python
ds2
```

```python
ds2.iloc[0:15].plot.pie(y='alcohol',figsize=(5,10),autopct='%.2f')
#displaying the particular values by locking
```

```python
pd.plotting.scatter_matrix(ds2,figsize=(10,10))
```

```python
pd.plotting.scatter_matrix(ds2,figsize=(10,10),diagonal='kde')
#plotting the graph with kernel density estimator.
```
